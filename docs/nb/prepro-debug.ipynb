```python
import pandas as pd


desimodel_raw_tables = pd.read_html("https://desidatamodel.readthedocs.io/en/latest/DESI_SPECTRO_DATA/NIGHT/EXPID/desi-EXPID.html")
desimodel_raw_tables = [desimodel_raw_tables[i].fillna(-9999).applymap(lambda s: s.replace("’","").replace("‘","").replace("…","") if type(s)==str else s).applymap(lambda s: pd.to_numeric(s, errors="ignore", downcast="integer")) for i in range(len(desimodel_raw_tables))]
```

```python
# generate dummy observations

import itertools as it

dummy_obs = fits.HDUList()

hdu = fits.PrimaryHDU(header=fits.Header(cards=map(tuple, desimodel_raw_sps.drop(columns="Type").values.tolist())))
hdu.name = "SPS"

dummy_obs.append(hdu)

for sp, i in it.product("Z", range(1)):
    
    hdu = fits.PrimaryHDU(
        header=fits.Header(map(tuple, desimodel_raw_cam.drop(columns="Type").values.tolist())),
        data=np.zeros((2048,2048))
    )
    hdu.name = f"{sp}{i}"
    
    dummy_obs.append(hdu)

hdu = fits.BinTableHDU(header=fits.Header(cards=map(tuple, desimodel_raw_con.drop(columns="Type").values.tolist())))
hdu.name = "SPECTCONS"

dummy_obs.append(hdu)
dummy_obs.writeto(OBS_NAME, overwrite=True)

dummy_obs.info()
```
